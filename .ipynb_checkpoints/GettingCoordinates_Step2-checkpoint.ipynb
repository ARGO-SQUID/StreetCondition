## Code below uses google API to geolocate all the 311 service requests with no coordinate information.

## It uses the available locations regestered by the 311 department to locate the data.

### *Note : All geo coordinates found are approxmated values based on the available information


```python
import pandas as pd
import numpy as np
#import Psycopg
% pylab inline
```

```python
data = []
for i in range(5,10):
    data.append(pd.read_csv('UnknownLocation/street'+str(i)+'.csv'))
```

```python
d2=pd.read_json('street10to16.json')
```

```python
df=pd.concat(data)
```

```python
df=pd.concat([df,d2])
```

```python
#df1=pd.read_csv('notlocated.csv')
```

```python
#potential fields containing possible addresses
address_loc = ['incident_address','intersection_street_1','intersection_street_2','landmark'
              ,'incident_zip','school_address','school_zip','cross_street_1','cross_street_2','borough','unique_key']
```

```python
df1 = df[address_loc]
```

```python
google_maps_key = 'Iput-your-api-key-here'
```

```python
df1=df1.fillna(np.nan)
```

```python
import re
def funfunction(x):
    z=''
    b=''
    try:
        z=int(x.incident_zip)
    except ValueError:
        z=''
    try:
        b=x.borough
    except ValueError:
        b=''
    if x.first_valid_index() is None:
        return None
    elif type(x.intersection_street_1)!=float and (type(x.intersection_street_2))!=float:
        #if(z!=''):
        k= x.intersection_street_1+', '+ x.intersection_street_2,b,str(z),'NY' 
        d=[]
        for i in k:
            if i!='':
                d.append(i)
        k=reduce(lambda x,y:x+','+y,d)
        return k
    elif (type(x[x.first_valid_index()])!= float) or (x[x.first_valid_index()]!='Unspecified'):
        k=x[x.first_valid_index()],b,str(z),'NY'
        d=[]
        for i in k:
            if i!='':
                d.append(i)
        k=reduce(lambda x,y:x+','+y,d)
        return k
    else:
        return 0
df1['validloc']=df1.apply(funfunction,axis=1)
```

```python
##using valid locations identified to be used as an input for coordinates identifiation 
import urllib2
import json
def locfinder(x):
        #print x.strip().replace(' ', '+') 
        url ='https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s'%(x.strip().replace(' ', '+'),\
                                                                                google_maps_key)
        request = urllib2.urlopen(url)
        data = json.load(request)
        if data['status']=="OVER_QUERY_LIMIT":
            return 0
        elif data['status']=="OK":
            return (data['results'][0]['geometry']['location']['lat'],data['results'][0]['geometry']['location']['lng'])
        else:
            #recursion to break string search to find coordinates for broader value. Ex. if MULBERRY STREET, 
            #MOSCO ST,MANHATTAN,NY returns none then look for MOSCO ST,MANHATTAN,NY then MANHATTAN,NY and so on
            if x.find(',')!=-1:
                x=x[x.find(',')+1:]
                return locfinder(x)
            pass
```

```python
df1['geom'] = df1['validloc'].apply(locfinder)
```

```python
#Seperating Longitudes and Latitudes
df1['latitude'] = map(lambda x: x[0],df1.geom)
df1['longitude'] = map(lambda x:x[1],df1.geom)
```

```python
df2= df1.copy()
```
